# Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
data['Sentiment'] = data['Speech Text'].apply(lambda text: sid.polarity_scores(text)['compound'])
```

```python
data.head()
```

```python
# Set a threshold for sentiment scores to identify outliers
sentiment_threshold = -0.3  # Adjust this threshold as needed

# Create a scatter plot for all points
fig = px.scatter(data, x="Year", y="Sentiment", color="Sentiment",
                 title="Sentiments of the Speeches Over Time",hover_name='Speaker')

# Filter data to include only points with sentiment scores exceeding the threshold
outliers = data[data['Sentiment'] < sentiment_threshold]

# Create a separate scatter plot for outliers with labels
fig.add_trace(px.scatter(outliers, x="Year", y="Sentiment", text="Speaker",hover_name='Speaker').data[0])

# Customize the layout
fig.update_traces(textposition='top center')  # Adjust the position of text labels

fig.show()
```

Comparing with other libraries

```python
from transformers import pipeline

# Transformers with BERT (example model)
sentiment_analysis = pipeline("sentiment-analysis")

# Function to process text and get sentiment score
def get_sentiment_score(text):
    # Split the text into chunks
    max_chunk_length = 512
    chunks = [text[i:i + max_chunk_length] for i in range(0, len(text), max_chunk_length)]

    # Calculate sentiment score for each chunk
    scores = [sentiment_analysis(chunk)[0]['score'] for chunk in chunks]

    # Use the average score of all chunks
    average_score = sum(scores) / len(scores)
    return average_score

# Apply the function to the 'Speech Text' column
data['BERT_Score'] = data['Speech Text'].apply(get_sentiment_score)
```

```python
from textblob import TextBlob
# TextBlob
data['TextBlob_Score'] = data['Speech Text'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
#trying out another model
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")

def get_sentiment_score(text):
    # Split the text into chunks
    max_chunk_length = 512
    chunks = [text[i:i + max_chunk_length] for i in range(0, len(text), max_chunk_length)]

    # Calculate sentiment score for each chunk
    scores = [sentiment_analysis(chunk)[0]['score'] for chunk in chunks]

    # Use the average score of all chunks
    average_score = sum(scores) / len(scores)
    return average_score

# Apply the function to the 'Speech Text' column
data['roberta_large_english'] = data['Speech Text'].apply(get_sentiment_score)

```

```python
pd.melt(data, value_vars=['Sentiment', 'TextBlob_Score', 'BERT_Score','roberta_large_english'],
                  var_name='Library', value_name='Sentiment Score')
```

# Clustering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Tf idf
tfidf = TfidfVectorizer(
    min_df = 5,
    max_df = 0.95,
    max_features = 8000,
    stop_words = 'english'
)
tfidf.fit(data['Speech Text'])
text = tfidf.transform(data['Speech Text'])
```

```python
from sklearn.cluster import MiniBatchKMeans
from yellowbrick.cluster import KElbowVisualizer

# Assuming you have a sparse matrix 'text' containing the speech data

# Try different values for the hyperparameters
init_size = 1024
batch_size = 2048
random_state = 42

model = MiniBatchKMeans(init_size=init_size, batch_size=batch_size, random_state=random_state,n_init=3)
visualizer = KElbowVisualizer(model, k=(1, 12))

visualizer.fit(text.toarray())  # Fit the data to the visualizer
visualizer.show()
```

```python
KM=MiniBatchKMeans(n_clusters=5, init_size=1024, batch_size=2048, random_state=42,n_init=3)
clusters = KM.fit_predict(text)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering()
visualizer = KElbowVisualizer(model, k=(1,12))
visualizer.fit(text.toarray())        # Fit the data to the visualizer
visualizer.show()
```

```python
aglm=AgglomerativeClustering(n_clusters=4)
aglom_clusters=aglm.fit_predict(text.toarray())
```

Comparing the Results

```python
from sklearn.metrics import silhouette_score,calinski_harabasz_score
```

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

```python
silhouette_avg = silhouette_score(text, KM.labels_)
print(f"Silhouette Score For Mini Batch KMeans Clustering: {silhouette_avg}")
```

```python
silhouette_avg = silhouette_score(text, aglm.labels_)
print(f"Silhouette Score For Aglomerative Clustering: {silhouette_avg}")
```

Calinski-Harabasz Index (Variance Ratio Criterion): Compares the ratio of between-cluster variance to within-cluster variance. Higher values are better.

```python
print(f"calinski_harabasz Score For Mini Batch KMeans Clustering: {calinski_harabasz_score(text.toarray(), KM.labels_)}")
```

```python
print(f"calinski_harabasz Score For Aglomerative Clustering: {calinski_harabasz_score(text.toarray(), aglm.labels_)}")
```

Both of these tests suggest that the mini bacth k means works better on the given data.

```python
# top keywords based on their TFIDF score to see if we can spot any trends
def get_top_keywords(data, clusters, labels, n_terms):
    df = pd.DataFrame(data.todense()).groupby(clusters).mean()

    for i,r in df.iterrows():
        print('\nCluster {}'.format(i))
        print(','.join([labels[t] for t in np.argsort(r)[-n_terms:]]))
```

```python
get_top_keywords(text, aglom_clusters, tfidf.get_feature_names_out(), 20)
```

```python
get_top_keywords(text, clusters, tfidf.get_feature_names_out(), 20)
```

Since we already know that the text comes from one specific type of speech type. we can see that it is very hard to find any subclusters in this this specific category. Based on our test results as well as teh visualizations it is fair to conclude that there are no clearly discernible clusters present in this text data.
